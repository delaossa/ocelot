# Optics for High Time Resolution Measurements with TDS

This tutorial is motivated by a real task: improving the time resolution of current profile measurements using a Transverse Deflecting Structure (TDS) at the European XFEL (EuXFEL).

## A Bit of Simple Theory

The transverse trajectory of a particle in a beamline is given by:

$$
x(s) = A \sqrt{\beta_x(s)} \cos(\Phi_x(s) + \Phi_0)
$$

where:

- $ \beta_x(s) $ is the betatron function at position $s $,
- $ \Phi_x(s) = \int_{s_0}^{s} \frac{1}{\beta_x(s)} \, ds $ is the betatron phase,
- $ \Phi_x'(s) = \frac{1}{\beta_x(s)} $.

Taking the derivative of $ x(s) $ with respect to $ s $, we get:

$$
x'(s) = -\frac{A}{\sqrt{\beta_x(s)}} \left[ \alpha_x(s) \cos(\Phi_x(s) + \Phi_0) + \sin(\Phi_x(s) + \Phi_0) \right],
$$

where:

- $ \alpha_x(s) = -\frac{1}{2} \beta_x'(s) $.

### At the TDS Position

At the position of the TDS (let's take $ s_0 = 0 $):

- The particle receives a transverse angular kick $ x'(s_0) = x'_{\text{tds}} $,
- The transverse position is $ x(s_0) = 0 $.

From the expression for $x(s) $:

$$
0 = A \sqrt{\beta_x(s_0)} \cos(\Phi_0)
$$

This implies:

$$
\cos(\Phi_0) = 0 \quad \Rightarrow \quad \Phi_0 = \frac{\pi}{2}
$$

Substituting into the expression for $ x'(s) $, we get:

$$
x'_{\text{tds}} = -\frac{A}{\sqrt{\beta_x(s_{\text{tds}})}}
$$

Solving for $ A $:

$$
A = -x'_{\text{tds}} \sqrt{\beta_x(s_{\text{tds}})}
$$

### At The Screen Position

The transverse position at the screen is:

$$
x(s_{\text{scr}}) = A \sqrt{\beta_x(s_{\text{scr}})} \cos(\Delta\Phi_x(s_{\text{scr}}) + \Phi_0),
$$
where $\Delta\Phi_x(s_{\text{scr}})$ is phase advance between TDS and screen. 

Using $ A = -x'_{\text{tds}} \sqrt{\beta_x(s_{\text{tds}})} $ and $ \Phi_0 = \frac{\pi}{2} $:

$$
x(s_{\text{scr}}) = -x'_{\text{tds}} \sqrt{\beta_x(s_{\text{tds}}) \beta_x(s_{\text{scr}})} \cos\left(\Delta\Phi_x(s_{\text{scr}}) + \frac{\pi}{2}\right)
$$

Recalling that:

$$
\cos\left(\psi + \frac{\pi}{2}\right) = -\sin(\psi)
$$

we get the final result:

$$
x(s_{\text{scr}}) = x'_{\text{tds}} \sqrt{\beta_x(s_{\text{tds}}) \beta_x(s_{\text{scr}})} \sin(\Delta\Phi_x(s_{\text{scr}}))
$$

## Transverse Kick from the Deflecting Structure

The deflecting structure imparts a time-dependent transverse momentum to the beam, resulting in a kick $ \Delta x'(t) $ given by:

$$
\Delta x_{\text{tds}}'(t) = \frac{e V_0}{p c} \sin\left( \frac{2\pi c t}{\lambda} + \varphi \right) 
\approx \frac{e V_0}{p c} \left( \frac{2\pi c t}{\lambda} \cos \varphi + \sin \varphi \right)
$$

where:

- $ V_0 $ is the peak voltage of the deflecting cavity,
- $ p $ is the longitudinal momentum of the electron bunch,
- $ c $ is the speed of light,
- $ \lambda $ is the RF wavelength,
- $ \varphi $ is the RF phase (with $ \varphi = 0 $ at the zero-crossing),
- $ t $ is the arrival time of the electron relative to the bunch center.

The approximation assumes $ ct / \lambda \ll 1 $, allowing the sine function to be expanded to first order.

Now we can calculate the beam size of the streaked beam on the screen. For diagnostic purposes, we typically set the RF phase to zero: $ \varphi = 0 $.

The rms beam size on the screen is then:

$$
\sigma_x^{\text{scr}} = \frac{e V_0}{p c}  \frac{2\pi c \sigma_t}{\lambda}  \sqrt{\beta_x(s_{\text{tds}}) \beta_x(s_{\text{scr}})} \sin(\Delta\Phi_x(s_{\text{scr}}))
$$

where $ \sigma_t $ is the rms bunch duration and $ \Delta\Phi_x(s_{\text{scr}}) = \Phi_x(s_{\text{scr}}) - \Phi_x(s_{\text{tds}}) $ is the betatron phase advance between the TDS and the screen.

---

## Time Resolution of the TDS

### Streaking Factor / Calibration Factor

The **streaking factor** (also known as the **calibration factor**) is the ratio of the streaked beam size on the screen $ \sigma_x^{\text{scr}} $ to the bunch length $ c \sigma_t $:

$$
S = \frac{\sigma_x^{\text{scr}}}{c \sigma_t} = \frac{e V_0}{p c}  \frac{2\pi}{\lambda}  \sqrt{\beta_x(s_{\text{tds}}) \beta_x(s_{\text{scr}})} \sin(\Delta\Phi_x(s_{\text{scr}}))
$$

### Time Resolution

The **rms time resolution** $ R_t $ of a TDS is defined as the ratio of the unstreaked beam size on the screen to the streaking factor:

$$
R_t = \frac{\sigma_{x0}^{\text{scr}}}{c S}
$$

Assuming the unstreaked beam size is given by $ \sigma_{x0}^{\text{scr}} = \sqrt{\varepsilon_x \beta_x(s_{\text{scr}})} $, we get:

$$
R_t = \frac{\sqrt{\varepsilon_x \beta_x(s_{\text{scr}})}}{c \cdot \left( \frac{e V_0}{p c}  \frac{2\pi}{\lambda}  \sqrt{\beta_x(s_{\text{tds}}) \beta_x(s_{\text{scr}})} \sin(\Delta\Phi_x(s_{\text{scr}})) \right)}
$$

Simplifying:

$$
R_t = \frac{\sqrt{\varepsilon_x}}{\frac{e V_0}{p}  \frac{2\pi}{\lambda}  \sqrt{\beta_x(s_{\text{tds}})} \sin(\Delta\Phi_x(s_{\text{scr}}))}
$$

As we can see, the time resolution **does not depend** on the beta function at the screen position. It depends only on the beta function at the TDS location.

**Conclusion:** To improve time resolution, we should **increase the beta function at the TDS position**. Let's do it!

## Make a function to display nicely Twiss parameters in particular elements. it can be done in different waus but we will use pandas for simplicity 

```python

# List of elements where we want to see Twiss parameters
elements_for_comparision = {'TDS 429': l2.marker_tds_b2, "Scr 450": l2.otrb_450_b2,  "Scr 454": l2.otrb_454_b2, 'Scr 457': l2.otrb_457_b2, 'Scr 461': l2.otrb_461_b2, 'end': l3.match_l3}

# Attributes we want to compare
attributes = ['beta_x', 'beta_y', 'alpha_x', 'alpha_y', 'mux']

def table_update(lat, tws0, elements_for_comparision, attributes):
    # calculate Twiss
    tws = twiss(lat, tws0=tws0, attach2elem=True)
    
    # Build the table from tws list
    table = pd.DataFrame({name: [getattr(getattr(obj, "tws"), attr) for attr in attributes] for name, obj in elements_for_comparision.items()},
                     index=attributes)
    # make phase advance in degree
    table.loc['mux'] = (table.loc['mux'] - table.loc['mux', 'TDS 429'])*180/np.pi

    # add R12 elements into table 
    R12_values = copy.copy(elements_for_comparision)
    for key in R12_values:
        stop_elem = elements_for_comparision[key]
        _, R, _ = lat.transfer_maps(energy=2.4, start=l2.marker_tds_b2, stop=stop_elem)
        R12_values[key] = R[0, 1]
    table.loc['R12'] = R12_values
    return table
table = table_update(lat, tws_match_385, elements_for_comparision, attributes)
table
```
